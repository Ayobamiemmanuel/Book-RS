### Drop all the  thay never rate a book

```python
#drop movie with user less than 10
pivot.dropna(thresh = 10, axis=1, how='all', inplace =True)
pivot.head()
```

### Normalizing

```python
pivot_norm = pivot.apply(lambda x: x - np.nanmean(x), axis=1)
pivot_norm.head()
```

```python
#Fill NaN with 0
pivot_norm.fillna(0, inplace = True)
pivot_norm.head()
```

```python
#convert into dataframe to make it easier
item_sim_df = pd.DataFrame(cosine_similarity(pivot_norm, pivot_norm), index=pivot_norm.index, columns=pivot_norm.index)
```

```python
def get_similar_anime_for_others(book_name):
    if book_name not in pivot_norm.index:
        return None, None
    else:
        #sim_books = item_sim_df.sort_values(by=['original_title'], ascending=False).index[1:]
        sim_books = item_sim_df.sort_values(by=book_name, ascending=False).index[1:]
       # sim_score = item_sim_df[book_name]sim_score = item_sim_df.sort_values(by=book_name, ascending=False).loc[:, book_name].tolist()[1:]
        sim_score = item_sim_df.sort_values(by=book_name, ascending=False).loc[:, book_name].tolist()[1:]
        return sim_books, sim_score
```

```python
books, score = get_similar_anime_for_others(" A Monster Calls ")
for x, y in zip(books[:10], score[:10]):
    print("{} with similarity of {}".format(x, y))
```

```python
#predict the rating of book x by user y
def predict_rating(user_id, book_name, max_neighbor=10):
    books, score = get_similar_anime_for_others(book_name)
    book_arr = np.array([x for x in books])
    sim_arr = np.array([x for x in score])
    
    #select only the book that has already rated by user x
    filtering = pivot_norm[user_id].loc[book_arr] != 0
    
    #calculate the predicted score
    s = np.dot(sim_arr[filtering][:max_neighbor], pivot[user_id].loc[book_arr[filtering][:max_neighbor]]) / np.sum(sim_arr[filtering][:max_neighbor])
    
    return s
```
